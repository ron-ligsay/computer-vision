# Transformation
rotating or resizing

## Why transform images?
* preparing images for classification Machine Learning Models
* Optimization and compression of images
* save images with some proportion

### Rotating

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

```python
# Import image from data folder
shittzu = plt.imread('../../Photos/shittzu.jpg')
```

```python
from skimage.transform import rotate

# Rotate the image 90 degress clockwise
shittzu_rotated = rotate(shittzu, angle=90, resize=True)

# Show the original and rotated image
plot_comparison(shittzu, shittzu_rotated, 'Rotated 90 degress clockwise')
```

```python
# Rotate the image 90 degress anti-clockwise
shittzu_rotated = rotate(shittzu, angle=-90, resize=True)

# Show the original and rotated image
plot_comparison(shittzu, shittzu_rotated, 'Rotated anti-clockwise')
```

## Rescaling

### Downgrading

```python
from skimage.transform import rescale

# Rescale the image to be 4 times smaller
shittzu_rescaled = rescale(shittzu, scale=0.25) #mode='reflect' anti-aliasing=True, multichannel=True

# Show the original and rescaled image
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax[0].imshow(shittzu, cmap=plt.cm.gray)
ax[0].set_title('Original image', fontsize=10)
#ax[0].axis('off')
ax[1].imshow(shittzu_rescaled, cmap=plt.cm.gray)
ax[1].set_title('rescaled image', fontsize=10)
#ax[1].axis('off')
```

### Aliasing in digital images
aliasing is a distortion in digital images that occurs when a signal is sampled at a rate lower than the Nyquist rate. The result is that the signal appears to have a lower frequency than it actually does.

# Resizing

```python
from skimage.transform import resize

# hieght and width of the image
height = 200
width = 180

# Resize the image
shittzu_resized = resize(shittzu, (height, width), anti_aliasing=True) #mode='reflect'

# Show the original and resized image
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax[0].imshow(shittzu, cmap=plt.cm.gray)
ax[0].set_title('Original image', fontsize=10)
ax[1].imshow(shittzu_rescaled, cmap=plt.cm.gray)
ax[1].set_title('resize image', fontsize=10)
```

### Resizing proportionally

```python
# Set proportional height so its 4 times its size
height = shittzu.shape[0] / 4
width = shittzu.shape[1] / 4

# Resize the image
shittzu_resized = resize(shittzu, (height, width), anti_aliasing=True) #mode='reflect'

# Show the original and resized image
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax[0].imshow(shittzu, cmap=plt.cm.gray)
ax[0].set_title('Original image', fontsize=10)
ax[1].imshow(shittzu_rescaled, cmap=plt.cm.gray)
ax[1].set_title('resize image', fontsize=10)
```
