```python
from skimage import data
rocket_image = data.rocket()
```

### Visualizing images in the course

```python
# Import Visualization Library
import matplotlib.pyplot as plt
```

```python
# Display the image
plt.imshow(rocket_image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

### RGB Channels

```python
# Import the color module from skimage
from skimage import color

# Display images with different color channels
fig, ax = plt.subplots(1, 3, figsize=(20, 10))

# Convert the image to different color spaces: gray, hsv, rgb
gray = color.rgb2gray(rocket_image)
hsv = color.rgb2hsv(rocket_image)
rgb = color.gray2rgb(gray)

# Plot the images in each color space
ax[0].imshow(gray, cmap='gray')
ax[1].imshow(hsv)
ax[2].imshow(rgb)

# Add titles to the color spaces
ax[0].set_title('Grayscale')
ax[1].set_title('HSV')
ax[2].set_title('RGB')

# Remove the ticks and tick labels
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')


# Show the figure
plt.show()
```
