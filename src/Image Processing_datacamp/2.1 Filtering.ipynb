# Filters
* Enchancing an image
* Emphasize or remove features
* Smoothing
* Sharpening
* Edge Detection

```python
# Import numpy
import numpy as np

# Import matplotlib
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
```

```python
# Impmort module and function
from skimage.filters import sobel
```

```python
# Import image from data folder
coins = plt.imread('../../Photos/coins.jpg')

# Display image
plt.imshow(coins)
```

```python
# Apply edge detection filter
edge_sobel = sobel(coins)
```

```python
# Show original and resulting image to compare
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4),sharex=True, sharey=True)

ax[0].imshow(coins, cmap=plt.cm.gray)
ax[0].set_title('Original image', fontsize=20)
ax[0].axis('off')

ax[1].imshow(edge_sobel, cmap=plt.cm.gray)
ax[1].set_title('Edge image', fontsize=20)
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

```python
plot_comparison(coins, edge_sobel, 'Edge with Sobel')
```

# Guassian Smoothing

```python
# Import image from data folder
shittzu = plt.imread('../../Photos/shittzu.jpg')

# Display image
plt.imshow(shittzu)
```

```python
# Import the module and function
from skimage.filters import gaussian

# Apply edge detection filter
gaussian_imamge = gaussian(shittzu) # multichannel=True # if the image is colored

# Show original and resulting image to compare
plot_comparison(shittzu, gaussian_imamge, 'Blurred with Gaussian filter')
```
