# Contrast  Enchancement

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Import image from data folder
lung = plt.imread('../../Photos/lung.jpg')
low_contrast = plt.imread('../../Photos/low_contrast.png') 
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

```python
# Display the image
plt.imshow(lung)
```

```python
# Histogram
plt.hist(lung.ravel(), bins=256, range=(0, 256), fc='k', ec='k')
plt.show()
```

```python
# Display the image
plt.imshow(low_contrast)
```

```python
# Histogram
plt.hist(low_contrast.ravel(), bins=256, range=(0, 256), fc='k', ec='k')
plt.show()
```

# Enchance Contrast
* Contrast Stretching
* Histogram Equalization
* Adaptive Histogram Equalization

## Types
* Histogram Equalization
* Adaptive Histogram Equalization
* Contrast Limited Adaptive Histogram Equalization

```python
# import png image
moon = plt.imread('../../Photos/moon.png')
```

## Histogram Equalization

```python
# Importing exposure
from skimage import exposure

# Obtain the equalized image
image_eq = exposure.equalize_hist(moon)

# Show original and result
plot_comparison(moon, image_eq, 'Histogram equalization')
```

## Adaptive Equlization
* contrastive limited adaptive histogram equalization

CLAHE in scikit-image

```python
# Apply adaptive equalization
image_adapteq = exposure.equalize_adapthist(moon, clip_limit=0.03)

# Show original and result
plot_comparison(moon, image_adapteq, 'Adaptive equalization')
```
