## Find contours using scikit-immage
Preparing the image

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

```python
# Import image from data folder
coins = plt.imread('../../data/photos/coins.jpg')
```

### Transform the image to 2d grayscale

```python
from skimage import color

# Make the image grayscale
coins_gray = color.rgb2gray(coins)
plot_comparison(coins, coins_gray, 'Grayscale')
```

### Binarize the image

```python
# Import filters
from skimage import filters

#  Obtain the thresh value
thresh = filters.threshold_otsu(coins_gray)

# Apply thresholding
binary = coins_gray > thresh
```

### Find contours using scikit-image

```python
# Import the measure module
from skimage import measure

# Find contours at a constant value of 0.8
contours = measure.find_contours(binary, 0.8)

# Show the image with contours found
fig, ax = plt.subplots()
ax.imshow(binary, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

### Contrast level value
various 0 to 1
the closer to 1, the more contours will be found

### The steps to spotting contours

1. Make the image grayscale
2. Obtain the optimal thres value of the image
3. Apply thresholding
4. Find contours at a constant value of 0.8
