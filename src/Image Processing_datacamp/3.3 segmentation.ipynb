# Segmentation

Example: before a tumor can be removed, it must be located. This is done by segmenting the tumor from the rest of the image.
or before recognizing a face, it must be picked out of the background


separating foreground from background

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Import image from data folder
madrid = plt.imread('../../Photos/madrid.jpg')
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

##  Superpixels
A superpixel is a group of pixels that are similar to each other. They are used to simplify an image and make it easier to process. They are also used to segment an image.

### Benefits of superpixels
* more meaningful regions
* computation efficiency

### two types of segmentation
* supervised - specified threshold value
* unsupervised - no threshold provided

## Unsupervised segmentation
Simple Linear Iterative Clustering (SLIC)) is an unsupervised segmentation algorithm. It is based on k-means clustering. It is a fast algorithm that produces good results.

```python
# Import the modules
from skimage.segmentation import slic
from skimage.color import label2rgb

# Obtain the segmentation
segments_slic = slic(madrid)#n_segments = 400, compactness = 10, sigma = 1

# Put segments on top of original image to compare
segmented_image = label2rgb(segments_slic, madrid, kind='avg')

# Show the segmented image
plot_comparison(madrid, segmented_image, 'SLIC')
```

```python
# Obtain the segmentation with 300 segments
segments_slic = slic(madrid, n_segments = 300)#, compactness = 10, sigma = 1

# Put segments on top of original image to compare
segmented_image = label2rgb(segments_slic, madrid, kind='avg')

# Show the segmented image
plot_comparison(madrid, segmented_image, 'SLIC')
```
