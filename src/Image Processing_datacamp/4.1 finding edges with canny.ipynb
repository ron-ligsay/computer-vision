# Finding the edges with Canny

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt

# Import image from data folder
coins = plt.imread('../../data/photos/coins.jpg')
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('Original image', fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

```python
from skimage import color
from skimage.feature import canny
from imageio import imread

# Convert image to grayscale
#coins = imread(coins)[:,:,:3]
coins = color.rgb2gray(coins)#, 3, channel_axis=2

# Apply Canny detector
edges = canny(coins)

# Show resulting image
plot_comparison(coins, edges, 'Edges with Canny')
```

## Canny Edge detector

```python
# Apply Canny detector with a sigm1 of 0.5
canny_edges_0_5 = canny(coins, sigma=0.5)
canny_edges_1 = canny(coins, sigma=1)

# Show resulting image
plot_comparison(canny_edges_1, canny_edges_0_5, 'Edges with Canny, sigma=0.5')
```
