# Histogram

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Import image from data folder
madrid_image = plt.imread('../../Photos/madrid.jpg')

# Display image
plt.imshow(madrid_image)
```

### Applications of histograms
* Analysis
* Thresholding
* Brightness and contrast
* Equalization

```python
# Obtaining the red values of the image
red_values = madrid_image[:, :, 0]

# Obtaining the green values of the image
green_values = madrid_image[:, :, 1]

# Obtaining the blue values of the image
blue_values = madrid_image[:, :, 2]

fig, ax = plt.subplots(2, 2, figsize=(12, 6))

# Obtain the histogram
ax[0][0].hist(red_values.ravel(), bins=256, fc='r', ec='r')
ax[0][1].hist(green_values.ravel(), bins=256, fc='g', ec='g')
ax[1][0].hist(blue_values.ravel(), bins=256, fc='b', ec='b')

# Show the histogram
plt.show()
```
