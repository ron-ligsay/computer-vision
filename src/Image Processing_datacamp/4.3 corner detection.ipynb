# Corner Detection

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt

# Import image from data folder
image = plt.imread('../../data/photos/pointy_building.png')
```

```python
# Function for image comparison
def plot_comparison(original, filtered, title, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6,8), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title(title, fontsize=10)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered, fontsize=10)
    ax2.axis('off')
```

## Matching Corners
* Harris corner detector

```python
shrine = plt.imread('../../data/photos/jinja-shrine.jpg')
```

```python
from skimage.feature import corner_harris, corner_subpix, corner_peaks
from skimage import color

# Convert image to grayscale
shrine_gray = color.rgb2gray(shrine)


# Apply the Harries corner detector on the image
harries = corner_harris(shrine_gray)

# Display the image
plot_comparison(shrine_gray, harries, 'Original', 'Harries')
```

this shows the points that can be based for corners

```python
# Finds t he coordinates of th corners
coords = corner_peaks(harries, min_distance=30) # min_distance = pixels

print("A total of {} corners were detected".format(len(coords)))
```

```python
def show_image_with_corners(image, coords):
    # Display the image with corners
    plt.imshow(image, interpolation='nearest',cmap=plt.cm.gray)
    plt.plot(coords[:, 1], coords[:, 0], '+r', markersize=15)
    plt.axis('off')
    plt.show()
```
