# Images as NdArrays

```python
# Import numpy
import numpy as np

# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Import image from data folder
madrid_image = plt.imread('images/madrid.jpg')

print('Type:', type(madrid_image))
```

### Colors with Numpy

```python
# Obtaining the red values of the image
red_values = madrid_image[:, :, 0]

# Obtaining the green values of the image
green_values = madrid_image[:, :, 1]

# Obtaining the blue values of the image
blue_values = madrid_image[:, :, 2]

# Displaying all the values
fig, ax = plt.subplots(1, 4, figsize=(16, 8))

ax[0].imshow(madrid_image)
ax[1].imshow(red_values, cmap='Reds')
ax[2].imshow(green_values, cmap='Greens')
ax[3].imshow(blue_values, cmap='Blues')

ax[0].set_title('Original image')
ax[1].set_title('Red values')
ax[2].set_title('Green values')
ax[3].set_title('Blue values')

# Remove the ticks and tick labels
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
ax[3].axis('off')


plt.show()
```

```python
# Displaying all the values
fig, ax = plt.subplots(1, 4, figsize=(16, 8))

# Displaying the values
ax[0].imshow(madrid_image)
ax[1].imshow(red_values, cmap='gray')
ax[2].imshow(green_values, cmap='gray')
ax[3].imshow(blue_values, cmap='gray')

# Set the title of each subplot
ax[0].set_title('Original image')
ax[1].set_title('Red values')
ax[2].set_title('Green values')
ax[3].set_title('Blue values')

# Remove the ticks and tick labels
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
ax[3].axis('off')


plt.show()
```
